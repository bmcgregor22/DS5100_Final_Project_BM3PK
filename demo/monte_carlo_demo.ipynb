```python
from montecarlo import Die, Game, Analyzer

# create two 6 sided die
sides = [1,2,3,4,5,6]
myDie1 = Die(sides)
myDie2 = Die(sides)
myDie3 = Die(sides)
myDie4 = Die(sides)
myDie5 = Die(sides)
        
# create list of two test tie
test_die = [myDie1, myDie2, myDie3, myDie4, myDie5]
        
# initialize a game object of two die
test_game = Game(test_die)

#conduct the test
test_game.play(10)
test_results = test_game.show()
print("printing the test results\n"+ str(test_results))
```

```python
#results of game in narrow forma
n=test_game.show('N')
n
```

```python
#wide format
w=test_game.show()
w
```
